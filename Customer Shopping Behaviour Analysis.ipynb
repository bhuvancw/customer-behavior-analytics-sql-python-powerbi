```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age-group
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
# create column purchase frequency days

frequency_mapping = {
        'Fortnightly': 14,
        'Weekly': 7,
        'Annually':365,
        'Quarterly':90,
        'Bi-Weekly':14,
        'Monthly':30,
        'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
# Check if all the values of both columns are equal
print((df['discount_applied'] == df['promo_code_used']).value_counts())
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Deleting column as it is not required
df = df.drop('promo_code_used', axis = 1)
```
